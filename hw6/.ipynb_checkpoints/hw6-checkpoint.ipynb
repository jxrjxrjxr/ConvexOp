## 使用鲁棒线性判别给出分类平面

使用鲁棒线性判别对给定的两个点集求解最优分类平面，使用凸优化的方式建立对应数学模型后，使用cvxpy求解即可，可以使用可视化工具先对点集的分布
有一个先验的认识，以验证实验结果的正确性

首先，通过`scipy.io.loadmat`读取两个点集的$X, Y$坐标：

```python
from scipy.io import loadmat
X1 = loadmat('x1.mat')['x'][0, :]
X2 = loadmat('x2.mat')['x'][0, :]
Y1 = loadmat('y1.mat')['y'][0, :]
Y2 = loadmat('y2.mat')['y2'][0, :]
```

画出两个点集的大致图像，分别用蓝色和红色区分:

```python
import numpy as np
X = np.concatenate((X1, X2), axis=None)
Y = np.concatenate((Y1, Y2), axis=None)
print('X1 dimension: ', X1.shape[0])
print('X2 dimension: ', X2.shape[0])
c = ['blue'] * X1.shape[0] + ['red'] * Y1.shape[0]
import matplotlib.pyplot as plt
plt.figure()
plt.scatter(X, Y, c=c)
plt.show()
```

通过上述步骤我们知道，两个点集都有1000个点，并且他们之间是线性可分的（且分隔线大概为$Y=X-2.5$）。因此我们使用如下的凸优化模型建立鲁棒线性判别：
$$
\begin{align*}
\max\ &t
\\
s.t.\ a^Tx_i-b &\ge t, i=1,2,...,1000
\\
a^Ty_i-b &\le -t,i=1,2,...,1000
\\
\lVert a \rVert_2 &\leq 1
\end{align*}
$$
其中，$x_i,y_i$分别是点集$X,Y$中的第$i$个点，$t$为待求解的分类间隔的一半，$a\in \mathbb{R}^{2*1}, b\in \mathbb{R}$为分类平面的参数。这个问题是一个凸问题（线性目标、线性和二次不等式），所以可以用`cvxpy`直接进行求解：

```python
import cvxpy as cp
import numpy as np
a = cp.Variable((2, 1))
b = cp.Variable()
t = cp.Variable()
obj = cp.Maximize(t)
constraint = []
Xm = np.stack([X1, Y1], axis=0)
Ym = np.stack([X2, Y2], axis=0)
for i in range(Xm.shape[1]):
    constraint.append(a.T @ Xm[:, i] - b >= t)
for i in range(Ym.shape[1]):
    constraint.append(a.T @ Ym[:, i] - b <= -t)
constraint.append(cp.norm(a) <= 1)
prob = cp.Problem(obj, constraint)
prob.solve()
```

下面输出最大分类间隔及分类平面：

```python
print(f'Max margin is {2 * t.value: .3f}')
print(f'The separating hyperplane is Y = {-a.value[0].item() / a.value[1].item():.3f} * X - {-b.value.item() / a.value[1].item():.3f}')
```

结果与猜想吻合。画出图像以验证正确性：

```python
Xl = np.arange(0, 10, 1./1000)
Yl = (b.value - a.value[0] * Xl) / a.value[1]
plt.figure()
plt.scatter(X, Y, c=c)
plt.plot(Xl, Yl, c='green')
plt.show()
```

```python
结果正确。
```
